## Grid 0.5

```r
grid_05 = get(load("../1.Upscaling/upscaling_ba/ba_grid_05.Rdata"))
#grid_fba = get(load("../1.Upscaling/upscaling_fba/ fba_grid_5.Rdata"))
df_grid = get(load("../3.Datos generados/dataframes/df_grid_05.Rdata"))
```

```r
df_grid_conFs <- df_grid[df_grid$FsOrNot != 0,]
df_grid_Length1 <- df_grid_conFs[df_grid_conFs$Length == 1,]
dfPhase <- data.frame('FS'=df_grid_Length1$MainStart,'P'= df_grid_Length1$P)
#dfPhase

vector_minimos <- c()
vector_maximos <- c()
vector_means <- c()
vector_medianas <- c()
for (i in 1:12){
    vector_minimos <- c(vector_minimos,round( min(dfPhase[dfPhase$FS==i,]$P),digits = 3))
    vector_maximos <- c(vector_maximos,round(max(dfPhase[dfPhase$FS==i,]$P),digits = 3))
    vector_means <- c(vector_means, round(mean(dfPhase[dfPhase$FS==i,]$P),digits = 3))
    vector_medianas <- c(vector_medianas, round(median(dfPhase[dfPhase$FS==i,]$P),digits = 3))
}
dfPhase.st <- data.frame('meses' = seq(1,12),vector_minimos,vector_maximos,vector_means,vector_medianas)
dfPhase.st 
```

```r
SeassonalTiming_viz <- quantity2clim(quantity = df_grid$'P', what = 'Phase', ref.grid = grid_05, backperm = NULL)
spatialPlot(SeassonalTiming_viz, backdrop.theme = 'coastline')
```
