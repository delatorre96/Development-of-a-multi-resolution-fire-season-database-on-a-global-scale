```r
df.seriesTemporales_fba <- func.ToDataFrame(grid = fba_5, coordX = fba_5$xyCoords$x, coordY = fba_5$xyCoords$y, func = mean)
```

```r
func.areasPixelesToDataFrame <- function(areasPixeles){
    superficie_pixel <- c()
    for (x in areasPixeles$xyCoords$x){ 
        for (y in areasPixeles$xyCoords$y){
            areasPixeles_subgrid <- subsetGrid(grid = areasPixeles, lonLim  = x, latLim  = y)
            superficie_pixel <- c(superficie_pixel, areasPixeles_subgrid$Data)
        }
    }
    return((superficie_pixel))
}
```

```r
vector.superficiePorPixel <- func.areasPixelesToDataFrame(areasPixeles_5)
```

Opcion de scale/nomarlizar 1

Usando Area quemada/(FBA * SuperficieDelPixel) => (Area Quemada)/(Area Quemable)

```r
df.seriesTemporales_scaled <- data.frame()
for (row in 1:nrow(df.seriesTemporales)){
    if (any(0 != df.seriesTemporales[row,])){
        serie_scaled <- unlist(df.seriesTemporales[row,])/(unlist(df.seriesTemporales_fba[row,])*vector.superficiePorPixel[row])
        df.seriesTemporales_scaled <- rbind(df.seriesTemporales_scaled, serie_scaled)
            }else{
        serie_scaled <- rep(0,12)
        df.seriesTemporales_scaled <- rbind(df.seriesTemporales_scaled, serie_scaled)
    }
}
```

```r
df.seriesTemporales_scaled_sinCeros <- func.eliminarSeriesConCeros(df.seriesTemporales_scaled,1)
df.seriesTemporales_scaled_sinCeros[1:10,]
```
